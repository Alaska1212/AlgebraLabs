## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
from pprint import pprint
G = SymmetricGroup(4)
subgroups = G.subgroups()

by_order = {}
for el in subgroups: 
    n = el.cardinality()
    if n not in by_order: 
        by_order[n] = []   
    by_order[n].append(el) 

pprint(by_order, indent=2)
```

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
import random

n = random.randint(1, 100)
S = SymmetricGroup(n)
Sn = SymmetricGroup(n)
An = AlternatingGroup(n)

s1 = S.random_element()
s2 = S.random_element()


even1 = s1.sign() == 1
even2 = s2.sign() == 1

G = S.subgroup([s1, s2])

order = G.order()

print(f"Підстановка 1: {s1}")
print(f"Парність: {even1}")
print(f"Підстановка 2: {s2}")
print(f"Парність: {even2}")
print(f"Порядок породженої групи: {order}")

if An.order() == order:
    print(f"Група, породжена підстановками 1 та 2, дорівнює An")
else:
    print(f"Група, породжена підстановками 1 та 2, не дорівнює An")

if Sn.order() == order:
    print(f"Група, породжена підстановками 1 та 2, дорівнює Sn")
else:
    print(f"Група, породжена підстановками 1 та 2, не дорівнює Sn")
```
