Спрогнозуйте скільки нормальних підгруп має група $D_{470448}$ __без виводу всіх за допомогою Sage__. _Опишіть_ всі нормальні підгрупи групи $D_{470448}$.  

```python
# Ваш код тут 
def dihedral_group_orders(n):
    G = DihedralGroup(n)
    normal_subgroups_list = G.normal_subgroups()
    return [subgroup.order() for subgroup in normal_subgroups_list]

normal_orders_dict = {}
for n in range(3, 101):
    normal_orders_dict[n] = dihedral_group_orders(n)

for n, orders in normal_orders_dict.items():
    print(f"D_{n} normal subgroups' orders: {orders}")
```
